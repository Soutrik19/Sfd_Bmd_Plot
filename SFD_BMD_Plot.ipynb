```python
from sympy.physics.continuum_mechanics.beam import Beam  #Importing Beam module from Sympy
from sympy import symbols,plot  #Importing Symbols and Plot module

def ssbeam():   #Function for Simply Supported Beam
    
    L = float(input("Enter the length of the beam in m : ")) 
    n = int(input("Enter the number of loading points for concentrated loads :"))
    m = int(input("Enter the number of loading points for moments :"))     
    c = int(input("Press 1 to include uniformly distributed load and 0 for not including uniformly distributed load :"))
    
    E,I = symbols('E , I')
    R1,R2 = symbols('R1, R2')
    E = float(input("Enter the value of Elastic Modulus in Pa :"))
    I = float(input("Enter the value of Area Moment of Inertia in m^4 :"))
    b = Beam(L,E,I)  #Creating a Beam variable 
    
    if(c==1):
        w = int(input("Enter 0 for Uniformly Distributed Load and 1 for Uniformly Varying load (Triangular) :"))
        UDL = float(input("Enter the value of uniformly distributed load in KN/m :"))
        p = float(input("Enter the distance from which intensity of UDL starts in m :"))
        q = float(input("Enter the distance upto which intensity of UDL remains in m:"))
        b.apply_load(UDL,p,w,end=q)  #Applying the UDL on beam
    else:
        print("Uniformly distributed load not included")
        
    P = []     #Declaration of Empty Lists
    M = []
    a1 = []
    a2 = []
            
    if(n>0):
        print("Enter the value(s) of the concentrated load(s) (in KN):")
        for i in range (0,n):
                  el = float(input())
                  P.append(el)  #Appending the concentrated load values in the List
            
        print("Enter the distance(s) at which the concentrated load(s) act (in m) :")
        for i in range (0,n):
                 el1 = float(input())
                 a1.append(el1) #Appending the values of distances at which the concentrated loads act in the List
    if(m>0):
        print("Enter the value(s) of the moment(s) (in KN.m):")
        for i in range (0,m):
                 el2 = float(input())
                 M.append(el2)  #Appending the values of Moments in the List
            
        print("Enter the distance(s) at which the moment(s) act (in m):")
        for i in range (0,m):
                 el3 = float(input())
                 a2.append(el3) #Appending the values of distances at which the moments act in the List
    else:
        print("No moments are applied for the loading configuration")
    
    b.apply_load(R1,0,-1)
    b.apply_load(R2,L,-1)
    b.bc_deflection.append((0,0))
    b.bc_deflection.append((L,0))
            
    for i in range (0,n):
             b.apply_load(P[i],a1[i],-1)  #Applying concentrated loads at entered points
    for i in range (0,m):
             b.apply_load(M[i],a2[i],-2)  #Applying Moments at entered points
            
    b.solve_for_reaction_loads(R1,R2)
    b.reaction_loads
    print("\n\nThe Free Body Diagram of the given problem is shown below :")
    h=b.draw()   #Implementing FBD graphics 
    h.show()
    print("\nThe Shear Force Diagram is plotted below :\n")
    b.plot_shear_force()
    print("\nThe Bending Moment Diagram is plotted below :\n")
    b.plot_bending_moment()
    print("\nThe point of maximum bending moment and the value of maximum bending moment are :")
    print(b.max_bmoment())
    print("\nThe Deflection Curve is plotted below :")
    b.plot_deflection()  

def cantbeam():  #Function for Cantilever Beam
    
    L = float(input("Enter the length of the beam in m : "))
    n = int(input("Enter the number of loading points for concentrated loads :"))
    m = int(input("Enter the number of loading points for moments :")) 
    c = int(input("Press 1 to include uniformly distributed load and 0 for not including uniformly distributed load :"))
    
    E,I = symbols('E , I')
    R,M = symbols('R , M')
    E = float(input("Enter the value of Elastic Modulus in Pa :"))
    I = float(input("Enter the value of Area Moment of Inertia in m^4 :"))
    b = Beam(L,E,I)  #Creating a Beam variable 
    
    if(c==1):
        w = int(input("Enter 0 for Uniformly Distributed Load and 1 for Uniformly Varying load (Triangular) :"))
        UDL = float(input("Enter the value of uniformly distributed load in KN/m :"))
        p = float(input("Enter the distance from which intensity of UDL starts in m :"))
        q = float(input("Enter the distance upto which intensity of UDL remains in m :"))
        b.apply_load(UDL,p,w,end=q)  #Applying the UDL on beam
    else: 
        print("Uniformly distributed load not included")
    
    P = []       #Declaration of Empty Lists
    Mo = []
    a1 = []
    a2 = []
    
    if(n>0):
        print("Enter the value(s) of the concentrated load(s) (in KN):")
        for i in range (0,n):
                  el = float(input())
                  P.append(el)   #Appending the concentrated load values in the List
            
        print("Enter the distance(s) at which the concentrated load(s) act (in m) :")
        for i in range (0,n):
                 el1 = float(input())
                 a1.append(el1)  #Appending the values of distances at which the concentrated loads act in the List
     
    if(m>0):
        print("Enter the value(s) of the moment(s) (in KN.m):")
        for i in range (0,m):
                 el2 = float(input())
                 Mo.append(el2)  #Appending the values of Moments in the List
            
        print("Enter the distance(s) at which the moment(s) act (in m):")
        for i in range (0,m):
                 el3 = float(input())
                 a2.append(el3)  #Appending the values of distances at which the moments act in the List
    else: 
        print("No moments are applied for the loading configuration")
    
    
    b.apply_load(R,0,-1)
    b.apply_load(M,0,-2)
    b.bc_deflection.append((0,0))
    b.bc_slope.append((0,0))
    
    for i in range (0,n):
             b.apply_load(P[i],a1[i],-1)  #Applying concentrated loads at entered points
    for i in range (0,m):
             b.apply_load(Mo[i],a2[i],-2) #Applying Moments at entered points
            
    b.solve_for_reaction_loads(R,M)
    b.reaction_loads
    print("\n\nThe Free Body Diagram of the given problem is shown below :")
    h=b.draw()    #Implementing FBD graphics 
    h.show()
    print("\nThe Shear Force Diagram is plotted below :\n")
    b.plot_shear_force()
    print("\nThe Bending Moment Diagram is plotted below :\n")
    b.plot_bending_moment()
    print("\nThe point of maximum bending moment and the value of maximum bending moment are :")
    print(b.max_bmoment())
    print("\nThe Deflection Curve is plotted below :")
    b.plot_deflection() 

    
def overbeam():  #Function for Overhanging Beam
    
    L = float(input("Enter the length of the beam in m : "))
    n = int(input("Enter the number of loading points for concentrated loads :"))
    m = int(input("Enter the number of loading points for moments :")) 
    a = float(input("Enter the distance of first support reaction from left end :"))
    t = float(input("Enter the distance of second support reaction from left end :"))
    c = int(input("Press 1 to include uniformly distributed load and 0 for not including uniformly distributed load :"))
    
    E,I = symbols('E , I')
    R1,R2 = symbols('R1, R2')
    E = float(input("Enter the value of Elastic Modulus in Pa :"))
    I = float(input("Enter the value of Area Moment of Inertia in m^4 :"))
    b = Beam(L,E,I)  #Creating a Beam variable 
    
    if(c==1):
        w = int(input("Enter 0 for Uniformly Distributed Load and 1 for Uniformly Varying load (Triangular) :"))
        UDL = float(input("Enter the value of uniformly distributed load in KN/m :"))
        p = float(input("Enter the distance from which intensity of UDL starts in m :"))
        q = float(input("Enter the distance upto which intensity of UDL remains in m :"))
        b.apply_load(UDL,p,w,end=q)  #Applying the UDL on beam
    else:
        print("Uniformly distributed load not included")
    
    P = []     #Declaration of Empty Lists
    M = []
    a1 = []
    a2 = []
    
    if(n>0):
        print("Enter the value(s) of the concentrated load(s) (in KN):")
        for i in range (0,n):
                  el = float(input())
                  P.append(el)  #Appending the concentrated load values in the List
            
        print("Enter the distance(s) at which the concentrated load(s) act (in m) :")
        for i in range (0,n):
                 el1 = float(input())
                 a1.append(el1)  #Appending the values of distances at which the concentrated loads act in the List
     
    if(m>0):
        print("Enter the values of the moment(s) (in KN.m):")
        for i in range (0,m):
                  el2 = float(input())
                  M.append(el2)  #Appending the values of Moments in the List
            
        print("Enter the distance(s) at which the moment(s) act (in m):")
        for i in range (0,m):
                  el3 = float(input())
                  a2.append(el3) #Appending the values of distances at which the moments act in the List
            
    
    b.apply_load(R1,a,-1)
    b.apply_load(R2,t,-1)
    b.bc_deflection.append((a,0))
    b.bc_deflection.append((t,0))
    
    for i in range (0,n):
             b.apply_load(P[i],a1[i],-1)  #Applying concentrated loads at entered points
    for i in range (0,m):
             b.apply_load(M[i],a2[i],-2)  #Applying Moments at entered points
     
    b.solve_for_reaction_loads(R1,R2)
    b.reaction_loads
    print("\n\nThe Free Body Diagram of the given problem is shown below :")
    h=b.draw()   #Implementing FBD graphics 
    h.show()
    print("\nThe Shear Force Diagram is plotted below :\n")
    b.plot_shear_force()
    print("\nThe Bending Moment Diagram is plotted below :\n")
    b.plot_bending_moment()
    print("\nThe point of maximum bending moment and the value of maximum bending moment are :")
    print(b.max_bmoment())
    print("\nThe Deflection Curve is plotted below :")
    b.plot_deflection() 

    
print("Given below are the choices of statically determinate beams for which SFD and BMD needs to be plotted:-") 
print("NOTE : Positive loads are taken downward and positive moments are taken anti-clockwise")  #for concentrated loads and moments
print("Enter 1 for simply supported beam")
print("Enter 2 for cantilever beam")
print("Enter 3 for overhanging beam")
ch = int(input("Enter the choice of statically determinate beam :")) #Choice of the beam type to be studied

if(ch==1):
    ssbeam()
elif(ch==2):
    cantbeam()
elif(ch==3):
    overbeam()
        
```
